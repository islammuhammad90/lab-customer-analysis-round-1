```python
import pandas as pd
import numpy as np
```

```python
# Reading data into a DataFrame

file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
# Displaying the shape of the 3 dataFrames

display(file1.shape)
display(file2.shape)
display(file3.shape)
```

```python
#Displaying the columns of each dataFrame

display(file1.columns)
display(file2.columns)
display(file3.columns)
display(file1.columns == file2.columns)
```

```python
# Displaying the head of the 3 DataFrames

display(file1.head())
display(file2.head())
display(file3.head())
```

```python
#Rename and Rearrange column names of file1

file1 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

file1 = file1[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#Rename and Rearrange column names of file2

file2 = file2.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

file2 = file2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#change CLV column to numeric in file 1

file1.dtypes
file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'].str.rstrip("%").astype(float)/100
```

```python
#change CLV column to numeric in file 2

file2.dtypes
file2['Customer Lifetime Value'] = file2['Customer Lifetime Value'].str.rstrip("%").astype(float)/100
```

```python
#Rearrange column names of file3

file3 = file3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
# concatenate the 3 files into 1 dataframe
data = pd.concat([file1,file2,file3], axis=0)
```

```python
# Displaying the shape of the newely created data frame

data.shape
```

```python
#Reading the top rows of the new DataFrame
data.head()
```

```python
#Droping the Education and n. of Open Complaints columns from the DataFrame

data = data.drop(['Education','Number of Open Complaints'], axis = 1)
```

```python
#Reading again the top rows after dropping the two columns above
data.head()
```

```python
#Checing each column data type
data.dtypes
```

```python
#Dropping duplicate rows from the data
data = data.drop_duplicates()
```

```python
#Checking the number of rows and columns after the dropped duplicate rows and requested columns
data.shape
```
