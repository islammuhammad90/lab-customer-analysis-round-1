```python
import pandas as pd
import numpy as np
```

```python
# Reading data into a DataFrame
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
#rearrange and rename column names of file1
file1 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

file1 = file1[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#rearrange and rename column names of file2
file2 = file2.rename(columns={'ST':'State100
                            'GENDER':'Gender'})

file2 = file2[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
#change CLV column to numeric in file 1

file1.dtypes
file1['Customer Lifetime Value'] = file1['Customer Lifetime Value'].str.rstrip("%").astype(float)/100
```

```python
#change CLV column to numeric in file 2

file2.dtypes
file2['Customer Lifetime Value'] = file2['Customer Lifetime Value'].str.rstrip("%").astype(float)/100
```

```python
#rearrange column names of file3
file3 = file3.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

file3 = file3[['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value','Income', 'Monthly Premium Auto', 'Number of Open Complaints','Policy Type', 'Vehicle Class', 'Total Claim Amount']]
```

```python
# concatenate the 3 files into 1 dataframe
data = pd.concat([file1,file2,file3], axis=0)
```
